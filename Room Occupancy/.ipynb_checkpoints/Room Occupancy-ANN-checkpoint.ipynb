## Data Visualization

```python
def histplot_and_boxplot(feature):
    plt.figure(figsize = (20,5))

    plt.subplot(1,2,1)
    ax = sns.histplot(x = feature,data = df,hue = 'Occupancy')
    ax.bar_label(ax.containers[0],size = 13)
    plt.xlabel(feature.title(),size = 16)
    plt.xticks(size = 14)
    plt.ylabel('Count',size = 16)
    plt.yticks(size = 14)
    plt.title('Distribution of '+feature.title(),size = 18,y = 1.05)

    plt.subplot(1,2,2)
    sns.boxplot(x = 'Occupancy',y = feature,data = df,palette = 'seismic')
    plt.xlabel('Occupancy',size = 16)
    plt.xticks(size = 14)
    plt.ylabel(feature,size = 16)
    plt.yticks(size = 14)

    plt.subplots_adjust(left = 0.1,right=0.9,wspace=0.4,hspace=0.4)
    plt.show()
```

```python
def jointplot(x_feature,y_feature):
    sns.jointplot(x = x_feature,y = y_feature,data = df,hue = 'Occupancy',palette = 'Set1')
    plt.title(y_feature.title()+' - '+x_feature.title(),size = 18,y = 1.3,x = -3)
```

## Data Preprocessing

### Removing Some Data

```python
df = df[df.Light<=1000]
```

### Creating the Independent and Dependent Variables

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

### Split the Data into the Train Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

### Scale the Variables

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Creating Artificial Neural Network Model

### Libraries

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import plot_model
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score
```

### Creating the ANN Model

```python
model = Sequential()

model.add(Dense(units=5,activation='relu'))
model.add(Dropout(rate = 0.4))

model.add(Dense(units=5,activation='relu'))
model.add(Dropout(rate = 0.4))

model.add(Dense(units=1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
early_stopping = EarlyStopping(mode = 'min',verbose=1,patience=10)
```

```python
model.fit(x=X_train,y = y_train,batch_size=32,epochs=100,validation_data=(X_test,y_test),callbacks=[early_stopping])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

## Plotting the Model

```python
plot_model(model,to_file='model.png')
```

## Result

```python
y_pred = (model.predict(X_test) > 0.5).astype('int32')
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
acc = round(accuracy_score(y_test,y_pred),3)
precision = round(precision_score(y_test,y_pred),3)
recall = round(recall_score(y_test,y_pred),3)

print(f'accuracy: {acc}\nprecision: {precision}\nrecall: {recall}')
```
